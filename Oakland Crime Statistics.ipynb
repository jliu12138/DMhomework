```python
from matplotlib.pyplot import *
import pandas as pd
import numpy as np
from scipy import stats

data = pd.read_csv('/home/kesci/input/DMhw_OCS12978/records-for-2011.csv')
```

```python
#去掉重复行
data.drop_duplicates()
#  标称属性
bc=['Agency','Area Id','Beat','Priority',\
    'Incident Type Id','Incident Type Description']
def plotbc(data,attribute):
    #去除空值
    data_attr=data[attribute].drop(data[data[attribute].isnull()].index)
    total = data[attribute].isnull().sum() #输出每个列缺失值也即值为NaN的数据和
    print(attribute +"缺失的条数为",total)
    figure(figsize=(15, 5))
    #统计某一列中各个元素值出现的次数
    x=data_attr.value_counts()  
    bar(x.index,x)
    show()
    #由于图上文字覆盖，打印频数
    print(x)
```

```python
plotbc(data,bc[0])
```

```python
#其余只缺失1条，直接剔除
data_new=data.drop(data[data['Agency'].isnull()].index)
#Area Id 和 Beat 缺失较多，用高频值填补这两个字段的缺失
mode=stats.mode(data_new['Area Id'])[0][0]
data_new['Area Id'] = data_new['Area Id'].fillna(mode)
mode=stats.mode(data_new['Beat'])[0][0]
data_new['Beat'] = data_new['Beat'].fillna(mode)
```
