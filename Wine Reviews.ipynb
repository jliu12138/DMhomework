```python
from matplotlib.pyplot import *
import pandas as pd
import numpy as np
from scipy import stats

data = pd.read_csv('/home/kesci/input/DMhw_WR2968/winemag-data_first150k.csv')
```

```python
#去掉重复行
data.drop_duplicates()
#  标称属性
bc=['country','province','region_1','region_2','variety']
def plotbc(data,attribute):
    #输出缺失个数
    data_attr=data[attribute].drop(data[data[attribute].isnull()].index)
    total = data[attribute].isnull().sum() #输出每个列丢失值也即值为NaN的数据和
    print(attribute +"丢失的条数为",total)
    figure(figsize=(15, 5))
    #统计某一列中各个元素值出现的次数
    x=data_attr.value_counts()  
    bar(x.index,x)
    show()
    #由于图上文字覆盖，打印频数
    print(x)
```

```python
plotbc(data,bc[0])
```

```python
#数值属性
sz=['points','price']
def plotsz(data,attribute):
    data_attr=data[attribute].drop(data[data[attribute].isnull()].index)
	#输出缺失个数
    total = data[attribute].isnull().sum() #输出每个列丢失值也即值为NaN的数据和
    print(attribute +"丢失的条数为",total)
    
    #求五数概括
    Minimum=min(data_attr)
    Q1=np.percentile(data_attr,25)
    Median=np.median(data_attr)
    Q3=np.percentile(data_attr,75)
    Maximum=max(data_attr)
    print(attribute +"的五数概括为：Minimum=", Minimum," Q1=", Q1, " Median=", Median," Q3=", Q3," Maximum=", Maximum)
    
    #绘图
    figure(figsize=(10, 30))
    boxplot(data_attr)
    show()
```

```python
plotsz(data,sz[0])
```

```python
plotsz(data,sz[1])
```

```python
#处理缺失数据
#缺失country的数据剔除
data_new=data.drop(data[data['country'].isnull()].index)
#缺失province的用高频值填充
mode=stats.mode(data_new['province'])[0][0]
data_new['province'] = data_new['province'].fillna(mode)

#缺失的用price的平均值填充
data_new['price'] = data_new['price'].fillna(data_new['price'].mean())
```

```python
plotbc(data_new,'country')
```

```python
plotbc(data_new,'province')
```

```python
plotsz(data_new,'price')
```
